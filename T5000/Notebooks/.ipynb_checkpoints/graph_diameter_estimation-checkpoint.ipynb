```python
x1 = sample.x
x0 = torch.rand(5000, 3)
t = torch.tensor([0.5]).view(-1, 1).expand(5000, 1)
x_t = min_image(x1-x0, **BOX) * t + x0
batch = torch.zeros(5000, dtype=int)

edge_index = radius_graph_pbc_batch(x_t, 0.3, batch, device)
G = nx.Graph()
G.add_edges_from(edge_index.T.tolist())
diameter = nx.diameter(G)
print(f"Diameter: {diameter}")
```
